```python
# Function to replace 
char_to_removes = ["+", ",", "$"]
column_to_clean = ["Installs", "Price"]

for item in char_to_removes:
    for column in column_to_clean:
        data_copy[column] = data_copy[column].str.replace(item, "")
```

```python
# Changing datatype form object to integr
data_copy["Installs"] = data_copy["Installs"].astype("int")
```

```python
# Changing datatype form object to float
data_copy["Price"] = data_copy["Price"].astype("float")
```

```python
# Converting Date which is written as str type now convert it into the data and time dtype using pands.
```

```python
data_copy["Last Updated"].dtype
```

```python
data_copy["Last Updated"]= pd.to_datetime(data_copy["Last Updated"])
```

```python
data_copy.head()
```

```python
data_copy.info()
```

```python
data_copy.head(3)
```

```python
#getting only date from this Lastupdated COlumns and create a new column of it
data_copy["Day"] = data_copy["Last Updated"].dt.day
```

```python
#getting only Month from this Lastupdated COlumns and create a new column of it
data_copy["Month"] = data_copy["Last Updated"].dt.month
```

```python
#getting only year from this Lastupdated COlumns and create a new column of it
data_copy["Year"] = data_copy["Last Updated"].dt.year
```

```python
data_copy.tail(20)
```

```python
text = [" and up", "Varies with device"]
for i in text:
    if i == "Varies with device":
        data_copy[""] = data_copy["Android Ver"].str.replace(i, str(np.nan))
    else:
        data_copy["Android Ver"] = data_copy["Android Ver"].str.replace(i, "")
```

```python
# Saving File 
data_copy.to_csv("Cleaned_Google_Playstore_Data", index=False)
```
